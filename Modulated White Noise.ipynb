```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
# Sampling rate
SR = 44100
t = np.linspace(0, 1, SR)

# White noise signal
x = 1.0 * np.random.rand(SR)

plt.figure()
plt.plot(t, x)
plt.title("Original White Noise")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Play original sound
sd.play(x, SR)
sd.wait()
```

```python
# Modulation parameters
fmod = 8
modlevel = 1
m = (modlevel / 2) * np.sin(2 * np.pi * fmod * t) + (1 - modlevel / 2)


plt.figure()
plt.plot(t, m)
plt.title("Modulation Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Apply modulation
y = x * m
plt.figure()
plt.plot(t, y)
plt.title("Modulated White Noise")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Play modulated sound
sd.play(y, SR)
sd.wait()
```
